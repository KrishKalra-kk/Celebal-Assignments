## Shopping Dataset Analysis

**Objective:** Load, explore, clean, and process a shopping dataset using Pandas.

## Part 1: Load and Setup

```python
import pandas as pd
import os

df = pd.read_csv('Combined_dataset.csv')
print("Dataset loaded successfully")
```

## Part 3: Handle Missing Values

```python
# Identify missing values
missing_count = df.isnull().sum()
print(missing_count[missing_count > 0])
```

```python
# Remove columns with too many missing values
df = df.dropna(axis=1, thresh=100)

# Fill remaining missing values
df = df.fillna('Unknown')

print(f"Missing values after cleaning: {df.isnull().sum().sum()}")
```

## Part 4: Select Relevant Columns

```python
# Select important columns
columns_to_keep = ['title', 'initial_price', 'final_price', 'rating', 'ratings_count', 'category']
df = df[columns_to_keep]

print(f"Shape after selecting columns: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

## Part 5: Remove Duplicates

```python
# Check duplicates
print(f"Duplicate rows before removal: {df.duplicated().sum()}")

# Remove duplicates
df = df.drop_duplicates()

print(f"Duplicate rows after removal: {df.duplicated().sum()}")
print(f"Dataset shape after removing duplicates: {df.shape}")
```

## Part 6: Create Derived Column

```python
# Convert prices to numeric
df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce')
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Create derived column: savings (discount amount)
df['savings'] = df['initial_price'] - df['final_price']

# Show the new column
print(df[['title', 'initial_price', 'final_price', 'savings']].head(10))
```

## Part 7: Basic Operations - Filter and Analyze

```python
# Filter: products with rating >= 4.0
high_rated = df[df['rating'] >= 4.0]
print(f"Products with rating >= 4.0: {len(high_rated)} out of {len(df)}")
print(f"Percentage: {(len(high_rated)/len(df)*100):.2f}%")
```

```python
# Select specific columns example
selected = df[['title', 'final_price', 'savings', 'category']]
print(selected.head(10))
```

## Part 8: Save Cleaned Dataset

```python
# Save cleaned dataset
df.to_csv('Combined_dataset_cleaned.csv', index=False)
print("Cleaned dataset saved as: Combined_dataset_cleaned.csv")
```
